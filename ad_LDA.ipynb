```python
import pandas as pd
```

```python
advertisement = pd.read_csv('/content/drive/MyDrive/BDP_final/oliveyoung_advertisement_preprocessed.csv', encoding='UTF-8')
```

```python
advertisement
```

```python
# 광고 문구 안에서만 볼거니까 ad만 남김.
advertisement_ad = advertisement.iloc[:, 4:]
advertisement_ad
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count = CountVectorizer(max_features=4000) # 자주 등장하는 단어 5000개로 제한
X = count.fit_transform(advertisement_ad['ad'].values)
```

```python
print(X)
```

```python
# LDA모델 만들고 fit

from sklearn.decomposition import LatentDirichletAllocation

lda = LatentDirichletAllocation(n_components=3, # 토픽 갯수
                                random_state=123, # 랜덤 시드
                                max_iter = 15, # 최대 반복횟수
                                learning_method='batch', # 학습 시킬 메서드
                                verbose=True) # verbose 매개변수 사용
X_topics = lda.fit_transform(X)
```

```python
# LDA 결과 토픽 보기

n_top_words = 20 # 토픽 관련 단어 20개 보기
feature_names = count.get_feature_names_out()

for topic_idx, topic in enumerate(lda.components_):
    print("Topic %d:" % (topic_idx + 1))
    print(" ".join([feature_names[i]
                    for i in topic.argsort()\
                        [:-n_top_words - 1:-1]]))
```

```python
import numpy as np
import pandas as pd
```

```python
smoothing_review = pd.read_csv('/content/drive/MyDrive/BDP_final/review_smoothing.csv', encoding='UTF-8')

# ê¸ì •ì¸ê²ƒë§Œ
smoothing_pos = smoothing_review[smoothing_review['soothing_pos'] == 1.0]

# ìƒí’ˆ ì´ë¦„ë³„ ê·¸ë£¹ë°”ì´, ê¸ì • ê°¯ìˆ˜
pos = smoothing_pos.groupby(['name']).count()

pos['soothing_pos'] = round(pos['soothing_pos'])

pos_sm = sum(pos['soothing_pos'])


#ë¶€ì •
smoothing_neg = smoothing_review[smoothing_review['soothing_neg'] == 1.0]

neg = smoothing_neg.groupby(['name']).count()

neg['soothing_neg'] = round(neg['soothing_neg'])

neg_sm = sum(neg['soothing_neg'])


#ê¸ì •/(ê¸ì •+ë¶€ì •)
print("smoothing_pos: " , pos_sm/(pos_sm + neg_sm))

#ë¶€ì •/(ê¸ì •+ë¶€ì •)
print("smoothing_neg: ", neg_sm/(pos_sm + neg_sm))

```

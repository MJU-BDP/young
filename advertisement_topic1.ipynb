```python
# 광고별 토픽 컬럼 추가 후 해당되는 토픽에 1 저장

# 경고 무시
import warnings
warnings.filterwarnings("ignore")

# LDA로 뽑은 토픽 단어사전
soothing = set(['진정', '자극', '트러블', '여드름', '민감','어성초', '티트리', '순하다'])
moisturizing = set(['보습', '흡수', '촉촉', '히알루론산', '건조' ])
skin = set(['피부결', '부드럽다', '정돈', '장벽', '탄력'])

topics = ['soothing', 'moisturizing', 'skin']

# dict = {'soothing': False, 'moisturizing': False, 'skin': False}

#column 추가 / 0으로 세팅
for wordlist_name in topics:
    advertisement["{}".format(wordlist_name)] = [0 for i in range(len(advertisement))]

# 광고에 토픽 단어 사전의 단어가 포함되면 해당 토픽으로 판단 (광고이니 모두 긍정)
for topic in topics:
  for keyword in eval(topic):
    for i in range(len(advertisement)):
      flag = True
      context = advertisement.iloc[i]['ad']
      if keyword in context and flag:
          advertisement["{}".format(topic)].iloc[i] = 1
          flag = False
```

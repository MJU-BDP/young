```python
import warnings
warnings.filterwarnings("ignore")

soothing_ = set(['진정', '자극', '트러블', '여드름', '민감','어성초', '티트리', '순하다'])
moisturizing = set(['보습', '흡수', '촉촉', ' 히알루론산', '건조' ])
skin = set(['피부결', '부드럽다', '정돈', '장벽', '탄력'])

topics = ['soothing', 'moisturizing', 'skin']

dict = {'soothing': False, 'moisturizing': False, 'skin': False}

#column 추가- soothing_pos,neg / moisturizing_pos,neg / skin_pos,neg
for wordlist_name in dict:
    advertisement["{}".format(wordlist_name)] = [0 for i in range(len(advertisement))]

for topic in topics:
    for keyword in eval(topic):
        for i in range(len(advertisement)):
            context = advertisement['ad']
            for text in context:
                if keyword in text and not dict[topic]:
                    advertisement.iloc[i]["{}".format(topic)] = 1
                else:
                    advertisement.iloc[i]["{}".format(topic)] = 1
```
